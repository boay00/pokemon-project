```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../../data/pokemon.csv")
```

```python
df.head()
```

```python
df = df[["attack", "defense", "sp_attack", "sp_defense", "speed"]]
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
features = [
    "attack",
    "defense",
    "sp_attack",
    "sp_defense"
    ]

target = "speed"
```

```python
X = df[features]
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

Optional Scaling (Not necessary for this specific task because the features are all in the same scale (pokemon stats are bound by a scale already :))

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)

X_train = pd.DataFrame(data = X_train_sc, columns = features)
X_test = pd.DataFrame(data = X_test_sc, columns = features)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
dummy_pokemon = {
    "name": "Bede's Dummy",
    "attack": 150,
    "defense": 79,
    "sp_attack":45,
    "sp_defense": 89
}
```

```python
dummy = pd.DataFrame(dummy_pokemon, columns = dummy_pokemon.keys(), index = range(1))
```

```python
dummy
```

```python
dummy_stats = dummy[features]
```

Since i scaled i now have to scale my dummy pokemon :)

```python
dummy_stats_sc = sc.transform(dummy_stats)
dummy_stats = pd.DataFrame(dummy_stats_sc, columns = dummy_stats.columns, index = dummy_stats.index)
```

```python
dummy_stats
```

```python
speed_stat = lr.predict(dummy_stats).item()
```

```python
speed_stat
```

```python
dummy_pokemon["speed"] = speed_stat
```

And finally heres my dummy pokemons new speed stat predicted by my model :)
